#### 1.简单版的TFIDF实现

```python
import jieba
import numpy as np
```

```python
documents = ["线程是程序执行时的最小单位，它是进程的一个执行流",
             "是CPU调度和分派的基本单位，一个进程可以由很多个线程组成",
        "线程间共享进程的所有资源，每个线程有自己的堆栈和局部变量。",
        "线程由CPU独立调度执行，在多CPU环境下就允许多个线程同时运行。",
        "同样多线程也可以实现并发操作，每个请求分配一个线程来处理。"]
```

```python
bagOfWords = [jieba.lcut(document) for document in documents]
```

```python
uniqueWords = set([w for words in bagOfWords for w in words])
```

```python
numOfWords = []
for bagOfWords_i in bagOfWords:
    numOfWords_i = dict.fromkeys(uniqueWords, 0)
    for word in bagOfWords_i:
        numOfWords_i[word] += 1
    numOfWords.append(numOfWords_i)
```

```python
def computeTF(wordDict, bagOfWords):
    tfDict = []
    for i in range(len(bagOfWords)):
        tfDict_i = {}
        bagOfWordsCount = len(bagOfWords[i])
        for word, count in wordDict[i].items():
            tfDict_i[word] = count / float(bagOfWordsCount)
        tfDict.append(tfDict_i)
    return tfDict
```

```python
tfs = computeTF(numOfWords, bagOfWords)
```

```python
def computeIDF(documents):
    import math
    N = len(documents)
    
    idfDict = dict.fromkeys(documents[0].keys(), 0)
    for document in documents:
        for word, val in document.items():
            if val > 0:
                idfDict[word] += 1
    
    for word, val in idfDict.items():
        idfDict[word] = math.log(N / float(val))
    return idfDict
```

```python
idfs = computeIDF(numOfWords)
```

```python
def computeTFIDF(tfs, idfs):
    tfidf = []
    for i in range(len(tfs)):
        tfidf_i = {}
        for word, val in tfs[i].items():
            tfidf_i[word] = val * idfs[word]
        tfidf.append(tfidf_i)
    return tfidf
```

```python
tfidf = computeTFIDF(tfs,idfs)
```

```python
import json
print(json.dumps(tfidf,indent=4,ensure_ascii=False))
```
